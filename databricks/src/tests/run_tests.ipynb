```python
import sys
import os

# Add the src directory to Python path so modules can be imported
src_dir = "/Workspace/Users/spyros.louvis@ctpsandbox.com/pub-python-repo/databricks/src"
if src_dir not in sys.path:
    sys.path.insert(0, src_dir)

test_dir = os.path.join(src_dir, "tests")
print(f"Source directory: {src_dir}")
print(f"Test directory: {test_dir}")
```

```python
import pytest

# Run all tests with verbose output
# -v: verbose, --tb=short: short traceback format
exit_code = pytest.main([
    test_dir,
    "-v",
    "--tb=short",
    "--no-header",
    "-q",
])

if exit_code == 0:
    print("\n✅ ALL TESTS PASSED")
else:
    print(f"\n❌ TESTS FAILED (exit code: {exit_code})")
```
